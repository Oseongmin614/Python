```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_csv("weight_horsepower_mpg.csv")
df
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="horsepower",
    y="mpg",
    data=df,
    alpha=0.5,
    color="blue")
plt.xlabel("horsepower")
plt.ylabel("fuel Efficiency(mpg)")

plt.title("horsepower vs Efficiency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weight=df["weight"]
weight
```

```python
horsepower = df["horsepower"]
horsepower
```

```python
mpg = df["mpg"]
mpg
```

```python
weight_mean = weight.mean()
weight_mean
```

```python
weight_std=weight.std()
weight_std
```

```python
weight_norm = (weight- weight_mean) / weight_std
weight_norm
```

```python
horsepower_mean=horsepower.mean()
horsepower_mean
```

```python
horsepower_std= horsepower.std()
horsepower_std
```

```python
horsepower_norm=(horsepower-horsepower_mean)/horsepower_std
horsepower_norm
```

```python
x1=weight_norm
x2=horsepower_norm
x2
```

```python
y= df["mpg"]
y
```

```python
a1=0.0
a2=0.0
b=0.0
```

```python
m=len(x1)
m
```

```python
epochs=5000
learning_rate=0.01
```

```python
for epoch in range(epochs):
    h= a1*x1+a2*x2+b
    error=h-y
    cost=(1/m)*np.sum(error**2)
    print(f"Epoch{epoch},cost:{cost:.4f},a1:{a1:.4f},a2:{a2:.4f},b:{b:.4f}")
    a1_grad=(1/m)*np.sum((h-y)*x1)
    a2_grad=(1/m)*np.sum((h-y)*x2)
    b_grad=(1/m)*np.sum(h-y)
    a1-=learning_rate*a1_grad
    a2-=learning_rate*a2_grad
    b -=learning_rate* b_grad
print(f"\n 최적화된 선형회귀식: mpg={a1:.4f} *weight+{a2:.4f} * horsepower+{b:.4f}")    
```

```python
weight_input=100
horsepower_input=120
weight_input_norm=(weight_input-weight_mean)/weight_std
weight_input_norm
horsepower_input_norm=(horsepower_input-horsepower_mean)/horsepower_std
horsepower_input_norm
```

```python
pp_mpg=a1*weight_input_norm+a2*horsepower_input_norm+b
print(f"\n weight=100, horsepower=120일때 연비:{pp_mpg:.4f}")
```
