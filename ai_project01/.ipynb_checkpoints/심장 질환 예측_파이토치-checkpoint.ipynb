```python
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
df = pd.read_csv('heart_disease.csv')
df
```

```python
number_columns=df.select_dtypes(include='number').columns.tolist()
number_columns
```

```python
len(number_columns)
```

```python
for col in number_columns:
    if col == "fbs":
        continue
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR = Q3 -Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
X= df.loc[:,"age":"thal"]
X
```

```python
Y= df[['target']]
Y
```

```python
X_mean = X.mean()
X_mean
```

```python
X_std = X.std()
X_std
```

```python
X_norm=(X-X_mean)/X_std
X_norm
```

```python
X_train, X_test,Y_train,Y_test=train_test_split(X_norm,Y,test_size=0.2, random_state=42)
```

```python
X_train_tensor=torch.tensor(X_train.values, dtype=torch.float32)
X_train_tensor
```

```python
Y_train_tensor=torch.tensor(Y_train.values,dtype=torch.float32)
Y_train_tensor[0:10]
```

```python
X_test_tensor= torch.tensor(X_test.values,dtype=torch.float32)
X_test_tensor
```

```python
Y_test_tensor=torch.tensor(Y_test.values,dtype=torch.float32)
Y_test_tensor
```

```python
input_dim = X_train.shape[1]
input_dim
```

```python
model=nn.Sequential(
    #1
    nn.Linear(input_dim,30),
    nn.BatchNorm1d(30),
    nn.ReLU(),
    nn.Dropout(0.5),
    #2
    nn.Linear(30,30),
    nn.BatchNorm1d(30),
    nn.ReLU(),
    nn.Dropout(0.5),
    #3
    nn.Linear(30,30),
    nn.BatchNorm1d(30),
    nn.ReLU(),
    nn.Dropout(0.5),

    #마지막 층
    nn.Linear(30,1),
    nn.Sigmoid()
)
```

```python
nn.init.kaiming_normal_(model[0].weight, nonlinearity='relu')
```

```python
nn.init.zeros_(model[0].bias)
```

```python
nn.init.kaiming_normal_(model[4].weight, nonlinearity='relu')
```

```python
nn.init.zeros_(model[4].bias)
```

```python
nn.init.kaiming_normal_(model[8].weight, nonlinearity='relu')
```

```python
nn.init.xavier_normal_(model[12].weight)
```

```python
nn.init.zeros_(model[12].bias)
```

```python
optimizer = optim.Adam(model.parameters(),lr=0.001)
optimizer
```

```python
loss_fn=torch.nn.BCELoss()
loss_fn
```

```python
BATCH_SIZE=64
train_dataset= TensorDataset(X_train_tensor,Y_train_tensor)
```

```python
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
epochs = 1000
for epoch in range(epochs):
    epoch_loss = 0.0
    for X_batch,Y_batch in train_loader:
        Y_pred=model(X_batch)
        loss = loss_fn(Y_pred, Y_batch)
        optimizer.zero_grad()
        loss.backward()
```
