```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df=pd.read_csv("weight_horsepower_mpg.csv")
x=df[["weight","horsepower"]]
x_std=x.std()
x_mean=x.mean()
x=(x-x_mean)/x_std
y=df[["mpg"]]
model=tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=[2])])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss='mean_squared_error')
```

```python
history =model.fit(x,y,epochs=100)
```

```python
a_values = model.layers[0].weights[0].numpy().flatten()
b_values = model.layers[0].weights[1].numpy()[0]
```

```python
print(f"학습된 기울기 a1(무게에 대한 계수): {a_values[0]:.4f}")
print(f"학습된 기울기 a2(무게에 대한 계수): {a_values[1]:.4f}")
print(f"학습된 절편 b: {b_values:.4f}")
```

```python
new_input= pd.DataFrame([[140,330]],columns=["weight","horsepower"])
new_input
```

```python
new_input = (new_input-x_mean)/x_std
new_input
```

```python
predicted_mpg=model.predict(new_input)
print(f"\n 예측된연비:{predicted_mpg}")
```
