```python
import numpy as np
```

```python
x= np.array([10, 7, 4, 0])

y = np.array([30, 50, 60, 80])
```

```python
a= -4.0
b=80.0
```

```python
m=len(x)

m
```

```python
#1번째 방법 값이 너무 길어진다. 

epochs = 100
for epoch in range(epochs):
    h = a*x + b
    error = h - y
    cost = (1/m) * np.sum(error**2)
    a_grad = (1/m) * np.sum(x * (h - y))
    b_grad = (1/m) * np.sum(h - y)
    a = a - a_grad
    b = b - b_grad
    print(f'[{epoch+1}/{epochs} epoch]')
    print(f'a (기울기): {a:.4f}')
    print(f'b (절편): {b:.4f}')
    print(f'cost: {cost:.4f}\n')
```

```python
epochs = 1000

lerning_rate =0.0001

for epoch in range(epochs):
    h=a*x+b
    error=h-y
    cost = (1/m)*np.sum(error**2)
    a_grad = (1/m) * np.sum(x * (h - y))
    b_grad = (1/m) * np.sum(h - y)
    a = a - a_grad*lerning_rate
    b = b - b_grad*lerning_rate
    print(f'[{epoch+1}/{epochs} epoch]')
    print(f'a (기울기): {a:.4f}')
    print(f'b (절편): {b:.4f}')
    print(f'cost: {cost:.4f}\n')
```
